# Cleaning data

```python
df=who[['Life expectancy ', 'Adult Mortality',
       'infant deaths', 'Alcohol', 'percentage expenditure',
        ' BMI ', 'under-five deaths ', 'Polio', 'Total expenditure',
       'Diphtheria ', 'Hepatitis B',' HIV/AIDS', 'GDP', 'Population',
       ' thinness  1-19 years', 
       'Income composition of resources', 'Schooling']]  #bmi,alcohol,diphtheria,Adult Mortality,HIV/AIDS,gdp,Schooling
df = df.dropna()
df = df.dropna(how='all',axis=1)

#df = df.fillna(method='ffill')
#df.shape
#df['GDP']=np.around(df['GDP'],decimals=3)

#print(df)
```

# Applying Regression

```python
X = df.drop('Life expectancy ', axis=1)
              
y = df['Life expectancy ']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state=10)
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train,y_train)
```

# Output

```python
print(lm.intercept_)
coeff_df = pd.DataFrame(lm.coef_,X.columns,columns=['Coefficient'])
coeff_df
```

# Modal Evaluation

```python
from sklearn import metrics
predictions = lm.predict(X_test)
plt.scatter(y_test,predictions)

print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
print('R^2:', metrics.r2_score(y_test, predictions))

#print(pd.DataFrame({'Actual': y_test, 'Predicted': predictions}))

```
